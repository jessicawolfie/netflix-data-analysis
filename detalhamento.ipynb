```python
import pandas as pd                
import matplotlib.pyplot as plt    
import seaborn as sns              
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
df.release_year.value_counts()
```

```python
movies = df[df['type'] == 'Movie'] 
```

```python
rel_year = movies.groupby('release_year').size().reset_index(name='movies_per_year')
```

```python
print(rel_year)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(
    data=rel_year,
    x='release_year',
    y='movies_per_year',
    palette='viridis'
)

plt.title('Number of films released per year', fontsize=16)
plt.xlabel('Release year')
plt.ylabel('Number of films')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
tv_show = df[df['type'] == 'TV Show'] 
```

```python
tvshow_rel_year = tv_show.groupby('release_year').size().reset_index(name='tv_show_per_year')
```

```python
print(tvshow_rel_year)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(
    data=tvshow_rel_year,
    x='release_year',
    y='tv_show_per_year',
    palette='viridis'
)

plt.title('Number of TV show released per year', fontsize=16)
plt.xlabel('Release year')
plt.ylabel('Number of TV show')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
